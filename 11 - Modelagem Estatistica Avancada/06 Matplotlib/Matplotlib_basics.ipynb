## First Touches

### Line Plot

Creating data for world's population *[billions]* in time *[years]*:

```python
pop = [2.519, 3.692, 5.263, 6.972]
year = [1950, 1970, 1990, 2010]
```

Getting help on functions (again, the best option is the documentation in https://matplotlib.org/api/pyplot_summary.html):

```python
#help(plt.plot)
#help(plt.plot()) #Note that this call gives an object explanation, not the function explanation.
```

Simple line plot is obtained by the *plt.plot* function:

```python
plt.plot(year, pop)
plt.show()
```

Adding title, axes labels and customizing axes ticks:

```python
plt.plot(year, pop)
plt.title("World Population Projections")
plt.xlabel("Year")
plt.ylabel("Population")
plt.yticks([0, 2, 4, 6, 8])
plt.show()
```

You can also customize axes ticks with text, as in y axis, to indicate the number is in *billions*:

```python
plt.plot(year, pop)
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("World Population Projections")
plt.yticks([0, 2, 4, 6, 8], ["0B", "2B", "4B", "6B", "8B"])
plt.show()
```

If you want to plot an area plot, coloring below the line, you can exchange *plt.plot* by *plt.fill_between*:

```python
plt.fill_between(year, pop, 0, color = "green") # The 0 means it will fill until the 0 on y axis.
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("World Population Projections")
plt.yticks([0, 2, 4, 6, 8], ["0B", "2B", "4B", "6B", "8B"])
plt.show()
```

You can also add markers to the line points, an customize markers and line:

```python
plt.plot(year, pop, color='red', marker='o', linestyle='dashed', 
         linewidth=2, markersize=12)
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("World Population Projections")
plt.yticks([0, 2, 4, 6, 8], ["0B", "2B", "4B", "6B", "8B"])
plt.show()
```

### Scatter Plot

With the same data (world's population *[billions]* in time *[years]*), the function to create a scatter plot is *plt.scatter*:

```python
plt.scatter(year, pop)
plt.show()
```

Again you can add title, axes labels and customizing axes ticks:

```python
plt.scatter(year, pop, color = 'red')
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("World Population Projections")
plt.yticks([0, 2, 4, 6, 8], ["0B", "2B", "4B", "6B", "8B"])
plt.show()
```

You can overplot two plots of different types, by adding plots (line plot + scatter plot) in sequence:

```python
plt.plot(year, pop)
plt.scatter(year, pop, color = 'red')
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("World Population Projections")
plt.yticks([0, 2, 4, 6, 8], ["0B", "2B", "4B", "6B", "8B"])
plt.show()
```

Now a new example, importing CSV data about programming language preferences.

Preparing libraries, style and importing CSV data:

```python
import csv
from collections import Counter

plt.style.use("fivethirtyeight")

data = pd.read_csv('data_progLangPref.csv')
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']
```

```python
data
```

```python
ids
```

```python
lang_responses
```

Counting and organizing responses:

```python
language_counter = Counter()

for response in lang_responses:
    language_counter.update(response.split(';'))
```

```python
language_counter
```

```python
languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
```

```python
languages, popularity
```

```python
languages.reverse()
popularity.reverse()
```

```python
languages, popularity
```

Plotting responses in a programming languages popularity chart:

```python
mpl.rcParams.update(mpl.rcParamsDefault) # Just to guarantee the style is set to matplotlib default.
plt.style.use("fivethirtyeight")

plt.barh(languages, popularity)

plt.title("Most Popular Languages")
# plt.ylabel("Programming Languages")
plt.xlabel("Number of People Who Use")

plt.tight_layout()

plt.show()
```

## Pie Plot

Let's suppose we know how many routines are functional in a set of companies for each programming language. Then we can visualize it in a pie plot:

```python
mpl.rcParams.update(mpl.rcParamsDefault) # Just to guarantee the style is set to matplotlib default.

plt.style.use("fivethirtyeight")

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0] # Percentage of the radius that piece will be exploded.

plt.pie(slices, labels=labels, explode=explode, shadow=True,
        startangle=90, autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'black'})

plt.title("Programming Language Popularity")
plt.tight_layout()
plt.show()
```

## Filling Area on Line Plots

Sometimes you not only want a line plot, but also want to colour the area below the line, or even between lines.

Importing data about the median income of developers by their most relevant programming language:

```python
data = pd.read_csv('data_devsMedianIncome.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

Before anything else, let's make a line plot with two lines, comparing Python developers with the overall developers community:

```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

# overall_median = 57287

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

If you want to know where Python developers median is above or below the overall median:

```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries > dev_salaries),
                 interpolate=True, alpha=0.25, label='Above Avg')

plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries <= dev_salaries),
                 interpolate=True, color='red', alpha=0.25, label='Below Avg')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

## Histogram

An automatic way to correctly format the date axis is with *plt.gcf().autofmt_xdate()*.

```python
from datetime import datetime, timedelta
from matplotlib import dates as mpl_dates

plt.style.use('seaborn')

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y, linestyle = 'solid')

plt.gcf().autofmt_xdate() # ADDED. gcf is 'get current figure' and the function autoformat the date values of the x axis.

plt.tight_layout()

plt.show()
```

Matplotlib own date formats can be used to format date axis:

```python
from datetime import datetime, timedelta
from matplotlib import dates as mpl_dates

plt.style.use('seaborn')

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y, linestyle = 'solid')

plt.gcf().autofmt_xdate() # gcf is 'get current figure' and the function autoformat the date values of the x axis.

date_format = mpl_dates.DateFormatter('%b, %d %Y') # ADDED. matplotlib has its own date formats.
plt.gca().xaxis.set_major_formatter(date_format) # ADDED. gca is 'get current axis' and the function set the formatter.

plt.tight_layout()

plt.show()
```

## Subplots

As we are doing until now, we are using plt object to plot. It holds standard plot conditions for a 1-graph plot:

```python
plt.style.use('seaborn')

data = pd.read_csv('data_devsMedianIncome.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, py_salaries, label='Python')
plt.plot(ages, js_salaries, label='JavaScript')

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

But there is a more organized way to plot. Not only more organized, but is the only way to produce multiple graphs and plot them at once. This way is called subplot, and it is compound by figures and axes.

A figure is the container holding our plottable objects (the whole window is shown when the code is ran).

An axis is a list of the actual matplotlib objects. A figure can contain multiple plots (axes).

Just transforming the previous plot in a subplot standard (just 1 axis):

```python
plt.style.use('seaborn')

data = pd.read_csv('data_devsMedianIncome.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig, ax = plt.subplots() # ADDED.

ax.plot(ages, py_salaries, label='Python') # CHANGED. plt. altered to ax.
ax.plot(ages, js_salaries, label='JavaScript') # CHANGED. plt. altered to ax.

ax.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs') # CHANGED. plt. altered to ax.

ax.legend() # CHANGED. plt. altered to ax.

ax.set_title('Median Salary (USD) by Age') # CHANGED. plt. altered to ax. When ax., as a standard, instead of title, 
#use set_title (I don't know why).
ax.set_xlabel('Ages') # CHANGED. plt. altered to ax. When ax., as a standard, instead of xlabel, 
#use set_xlabel (I don't know why).
ax.set_ylabel('Median Salary (USD)') # CHANGED. plt. altered to ax. When ax., as a standard, instead of ylabel, 
#use set_ylabel (I don't know why).

plt.tight_layout()

plt.show()
```

*plt.subplots()* creates as default 1 row and 1 column subplots (just one plot). Note that when we create subplots with more than 1 row and/or column, matplotlib objects are created and organized in lists:

```python
fig, ax = plt.subplots(nrows = 2, ncols = 1)
print(ax)
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2)
print(ax)
```

Then we can plot 2 subplots doing creating for the sabe figure two axis containing one plot each:

```python
plt.style.use('seaborn')

data = pd.read_csv('data_devsMedianIncome.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig, (ax1, ax2) = plt.subplots(nrows = 2, ncols = 1)

ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
ax1.legend() 
ax1.set_title('Median Salary (USD) by Age')
ax1.set_xlabel('Ages')
ax1.set_ylabel('Median Salary (USD)')

ax2.plot(ages, py_salaries, label='Python') 
ax2.plot(ages, js_salaries, label='JavaScript')
ax2.legend() 
ax2.set_title('Median Salary (USD) by Age')
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

We should not repeat the title in the figure above, so we can remove it from the second plot. And we can also make the x-axis to be the same for both plots. See it below:

```python
plt.style.use('seaborn')

data = pd.read_csv('data_devsMedianIncome.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig, (ax1, ax2) = plt.subplots(nrows = 2, ncols = 1, sharex = True) # CHANGED.

ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
ax1.legend() 
ax1.set_title('Median Salary (USD) by Age')
#ax1.set_xlabel('Ages') # CHANGED.
ax1.set_ylabel('Median Salary (USD)')

ax2.plot(ages, py_salaries, label='Python') 
ax2.plot(ages, js_salaries, label='JavaScript')
ax2.legend() 
#ax2.set_title('Median Salary (USD) by Age') # CHANGED.
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```
